```python
import glob
allimglist = sorted(glob.glob("./UFPR-ALPR dataset/*/*/*.png"))
allimglabels = sorted(glob.glob("./UFPR-ALPR dataset/*/*/*.txt"))
```

```python
from PIL import Image,ImageDraw
import shutil
```

```python
for i in allimglist:
    print(i)
```

```python
def convert(size, box):
    dw = 1.0/size[0]
    dh = 1.0/size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
for fileidx,filename in enumerate(allimglabels):
    print(fileidx,filename)
    
    im=Image.open(allimglist[fileidx])
    w= int(im.size[0])
    h= int(im.size[1])
    
    print(w,h,im)

    file1 = open(filename,"r+")
    content = file1.readlines()
    for line in content:
        if(line.find("position_plate") != -1):
            print(line)
            coordinates = tuple(line.split()[-4:])
            coordinates = [int(i) for i in coordinates]
            coordinates[2] += coordinates[0]
            coordinates[3] += coordinates[1]
            coordinates = [coordinates[0],coordinates[2],coordinates[1],coordinates[3]]
            print(coordinates)
            yolocoordinates = convert((w,h),coordinates)
            print(yolocoordinates)
    file1.seek(0)
    file1.truncate()
    yolocoordinates = list(yolocoordinates)
    file1.write("0 " + str(yolocoordinates[0]) + " " + str(yolocoordinates[1]) + " " + str(yolocoordinates[2]) + " " + str(yolocoordinates[3]))
    file1.close()
```
